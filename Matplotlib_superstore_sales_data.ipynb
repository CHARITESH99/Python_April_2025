## BAR CHART - Discrete data

```python
agg = df.groupby('Region')['Sales'].sum()
```

```python
agg
```

```python
x = agg.index
```

```python
y = agg.values
```

```python
x
```

```python
y
```

```python
plt.bar(x,y,color = ['red','blue','green','orange'],alpha = 0.5, hatch = '/')
```

## PIE CHART - category wise distribution, mention values first, next labels, Explode - highlights the portion, autopct - shows percentages

```python
agg
```

```python
labels = agg.index
```

```python
values = agg.values
```

```python
plt.pie(values,labels = labels,colors = ['red','green','yellow','b'],explode=(0.1,0,0.3,0),autopct='%1.2f%%')
```
