```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
data = pd.DataFrame(np.random.randint(0,100,(10,3)),columns=['sales','units','transactions'])
```

```python
data
```

```python
data['sales']
```

```python
plt.plot(data['sales'])
plt.xlabel('index')
plt.ylabel('sales')
plt.title('sales distribution')
```

## CUSTOMIZATIONS

### 1. Color

```python
plt.plot(data['sales'],color = 'r')
plt.xlabel('index')
plt.ylabel('sales')
plt.title('sales distribution')
```

### 2. Marker - points on the graph( we can have till 11, marker = 1 to 11)

```python
plt.plot(data['sales'],color = 'r',marker = 'o')
plt.xlabel('index')
plt.ylabel('sales')
plt.title('sales distribution')
```

### 3. Figure(figsize =(x-axis,y-axis)

```python
plt.figure(figsize=(15,5))
plt.plot(data['sales'],color = 'r',marker = 'o')
plt.xlabel('index')
plt.ylabel('sales')
plt.title('sales distribution')
```

### 4. linewidth - thickness of the line

```python
plt.figure(figsize=(15,5))
plt.plot(data['sales'],color = 'r',marker = 'o',linewidth = 3)
plt.xlabel('index')
plt.ylabel('sales')
plt.title('sales distribution')
```

### 5. alpha - saturation means thickness of the color, max = 1(thicker) , min = 0 

```python
plt.figure(figsize=(15,5))
plt.plot(data['sales'],color = 'r',marker = 'o',linewidth = 3,alpha =0.5)
plt.xlabel('index')
plt.ylabel('sales')
plt.title('sales distribution')
```

### 6. marker, markersize, markeredgecolor, markerfacecolor, markeredgewidth

```python
plt.figure(figsize=(15,5))
plt.plot(data['sales'],color = 'r',marker = 'o',markersize = 10,markeredgecolor = 'b', markerfacecolor = 'w',markeredgewidth = 2)
plt.xlabel('index')
plt.ylabel('sales')
plt.title('sales distribution')
```

### Font size, color, labelpad(gives space between axis and label), pad for title

```python
plt.plot(data['sales'],color = 'b',marker = 'o')
plt.xlabel('index',fontsize=10,color = 'green',labelpad = 0)
plt.ylabel('sales',fontsize=10,color = 'green',labelpad = 0)
plt.title('sales distribution',fontsize=10,color = 'green',pad = 0)
```

### Grid - lines present background, grid(False)  - off,  linestyle, linewidth, alpha

```python
plt.grid(linestyle = ':',linewidth=4,alpha = 0.5)
plt.plot(data['sales'],color = 'b',marker = 'o')
plt.xlabel('index',fontsize=10,color = 'green',labelpad = 0)
plt.ylabel('sales',fontsize=10,color = 'green',labelpad = 0)
plt.title('sales distribution',fontsize=10,color = 'green',pad = 0)
```

### Xticks - fontsize, rotation 

```python
plt.plot(data['sales'],color = 'b',marker = 'o')
plt.xticks(fontsize=20,rotation=45)
```

### Legend

```python
plt.plot(data['sales'],color = 'b',marker = 'o')
plt.xticks(fontsize=20,rotation=45)
plt.legend(loc='upper right')
```
