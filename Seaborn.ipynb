## Histogram using sns

```python
sns.histplot(df['Sales'])
```

## Box plot using sns

```python
plt.figure(figsize = (12,6))
sns.boxplot(df['Sales'])
```

```python
plt.figure(figsize = (12,6))
sns.boxplot(data = df, x = 'Category', y = 'Sales')
plt.title('Boxplot of Profit by Category')
plt.xticks(rotation = 45)
plt.show()
```

## Violin Plot using sns

```python
plt.figure(figsize = (12,6))
sns.violinplot(data = df, x = 'Category', y = 'Sales')
plt.xticks(rotation = 45)
plt.show()
```

## Count plot using sns

```python
plt.figure(figsize = (12,6))
sns.countplot(data = df, x = 'Region')
plt.title('Count Plot of Orders by Region')
plt.show()
```

```python
df['Region'].value_counts()
```

## Bar plot using sns, we can use estimator = np.sum, np.mean, np.std, np.var, np.count

```python
plt.figure(figsize  = (12,6))
sns.barplot(data = df, x = 'Ship Mode', y = 'Sales', estimator = np.sum)
plt.title('Bar Plot of Profit by Ship Mode')
```

## Line plot 

```python
plt.figure(figsize = (12,6))
sns.lineplot(data = df, x = 'Order Date', y = 'Sales')
plt.title('Line Plot of Sales over Time')
plt.show()
```

## Scatter Plot, hue = category means it gives different colors for different categories

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Category')
plt.title('Scatter Plot of Sales vs Profit by Category')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Sales', y='Profit')
plt.title('Scatter Plot of Sales vs Profit by Category')
plt.show()
```

## Heat Map

```python
plt.figure(figsize=(20,20))
sns.heatmap(df['Sales'].values[0:100].reshape(10,10),annot=True,cmap='viridis')
```

## Swarm Plot 

```python
sns.swarmplot(data=df.sample(1000), x='Category', y='Sales')
```

## Regression Plot , y = mx+c draws a line between two sides

```python
sns.regplot(data=df, x='Sales', y='Profit')
```

## LM Plot 

```python
sns.lmplot(data=df, x='Sales', y='Profit', hue='Category')
```
