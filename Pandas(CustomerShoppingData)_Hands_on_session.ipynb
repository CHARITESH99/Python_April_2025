### 8. Create age buckets (e.g., 0–20, 21–40, 41–60, 61+)

```python
def age_buckets(age_number):
    if age_number<=20:
        return '0-20'
    elif age_number<=40:
        return '21-40'
    elif age_number<=60:
        return '41-60'
    else:
        return '61+'
customer_shopping_data['age_buckets']=customer_shopping_data['age'].apply(age_buckets)
```

```python
customer_shopping_data.head()
```

### 9. Add a column is_weekend based on invoice_date.

```python
customer_shopping_data['day_name']=customer_shopping_data['invoice_date'].dt.day_name()
def is_weekend(day_name):
    if day_name in ['Saturday','Sunday']:
        return True
    else:
        return False
customer_shopping_data['is_weekend']=customer_shopping_data['day_name'].apply(is_weekend)
customer_shopping_data.head()
```
